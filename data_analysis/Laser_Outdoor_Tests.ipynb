# Laser Sensor Outdoor Tests

## Objective

The objective of this notebook is to determine the sensor's effectiveness in detecting and measuring the distances of vehicles in a real-world setting.

First, I mounted the sensor on a stationary bike and positioning it along a busy road to test its ability to detect passing vehicles. Next, with the same set up, the bike was cycled down different roads and the data from overtaking vehicles was collected.
```python
import pathlib

import matplotlib.pyplot as plt

from data import DataLoader
from graphing import OutdoorGraphs
from preprocessing import DataCleaner
```

## Stationary Test

In the first test, the sensor was mounted on a stationary bike along Yuan Ching Rd. which saw moderate traffic. I also mounted a Go Pro to capture a video of the vehicles that passed the bike. I cross referenced the data collected from the sensor and the Go Pro's video feed to determine the reliability of the sensor.

### Setup

- I first define the paths to my data and initialize my modules.
- I also create a helper function that loads the relevant data (times, distances measured, and signal strength) from a given data file, and filters out points above 3.5m

```python
STATIONARY_PATH_1 = pathlib.Path("./data/laser_outdoor_tests/yuan_ching_rd.txt")
STATIONARY_PATH_2 = pathlib.Path("./data/laser_outdoor_tests/yuan_ching_rd_2.txt")
FOOD_DELIVERY_1 = pathlib.Path("./data/laser_outdoor_tests/food_delivery_1.txt")

data_loader = DataLoader()
graphs = OutdoorGraphs()
data_cleaner = DataCleaner()
```

```python
def extract_data_from_file(file_path: pathlib.Path) -> tuple[list]:
    """
    Helper function that extracts data from file.

    Args:
        file_path (pathlib.Path): Data file.

    Returns:
        tuple[list]: Returns timings, distances, and signal strengths.
    """
    timings, distances, strengths = data_loader.load_data_from_file(file_path)
    distances = data_cleaner.fliter_above(distances, 3500)
    return timings, distances, strengths
```

### Raw Data

- Because the sensor measures up to 100m, I cleaned out data above 3.5m for the scatter plot below.
- Unlike the TOF sensor, the laser sensor barely produced any spurious data, and was a lot more accurate and precise.
- However, despite advertising its frequency as 20Hz, the actualy frequency was inconsistent and slow, averaging below 10Hz and was therefore unable to capture all the vehicles that passed. As such, even though points were grouped together, they are few and far between as seen from the graph below.

```python
timings, distances, strengths = extract_data_from_file(STATIONARY_PATH_1)
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
graphs.scatter_time_vs_distance(axes, timings, distances, title="Yuan Ching Rd Raw Data")
plt.show()
```

- The software provided offered a command that allowed the sensor to run consistently at 20Hz. Therefore, I temporarily switched to using the windows and tested the sensor on the same setup.
- Like above, I load the data and filter all distances above 3.5m.

```python
timings, distances, strengths = extract_data_from_file(STATIONARY_PATH_2)
```

- From the graph, we can see how the sensor collected many more points when it's frequency was higher.
- From the way the points are consistently clustered together, we can also infer it's reliability.

```python
fig, axes = plt.subplots(figsize=(12, 6))
graphs.scatter_time_vs_distance(axes, timings, distances, title="Yuan Ching Rd Raw Data 2")
plt.show()
```

- After analyzing the video feed and annotating the graph, I once again confirmed the graphs reliability for detecting vehicles. The sensor performs best at detecting slow or stationary vehicles, as seen by the dark line in the center that represents the stationary bus, or the 3 dark clusters of points just after 19:06:01 where 3 slow moving vehicles passed the sensor.
- However, around 19:06:40, the vehicles in the video are moving very quickly, and the sensor appears to have missed out on 5 of these vehicles.
